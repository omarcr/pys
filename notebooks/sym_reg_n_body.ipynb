```python
import sympy
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pysr import PySRRegressor
```

```python
df = pd.read_csv('/home/omar/Documents/Code/symbolic_deep_learning/nbody_data.csv')
df =df.head(12000)
print(df.head()) 
```

```python
data = df.to_numpy()
```

```python
y = df[df.columns[0]].to_numpy()
y
```

```python
X = df.iloc[:, 1:].to_numpy()
X
```

```python
default_pysr_params = dict(
    populations=30,
    model_selection="best", # OAC: Other options are "accuracy" and "score"
)
```

```python
# Learn equations
model = PySRRegressor(
    niterations=30,
    binary_operators=["plus", "-", "mult", "div"],
    #unary_operators=["cos", "exp", "sin"],
    **default_pysr_params
)

model.fit(X, y)
```
